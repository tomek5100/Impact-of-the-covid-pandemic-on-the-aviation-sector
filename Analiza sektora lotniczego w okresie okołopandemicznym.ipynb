Ilość przypadków w poszczególnych miesiącach

```python
USA_data = data['USA']['data']
# ilość nowych przypadków 

cases_by_month = {}

for element in data['USA']['data']:
  year = element['date'][0:4]
  month = element['date'][5:7]
  if year not in cases_by_month:
    cases_by_month[year]  = {}
  if month not in cases_by_month[year]:
    cases_by_month[year][month]  = 0
  if 'new_cases' in element:
    cases_by_month[year][month] += element['new_cases']

#print(cases_by_month)
df_usa_by_month = pd.DataFrame.from_dict(cases_by_month)
df_usa_by_month
```

```python
df_usa_by_month.to_excel("USA_by_month.xlsx")
```

Wszystkie przypadki miesiąc po miesiącu

```python
total_cases_by_month = {}

for element in data['USA']['data']:
  year = element['date'][0:4]
  month = element['date'][5:7]
  if year not in total_cases_by_month:
    total_cases_by_month[year]  = {}
  if month not in total_cases_by_month[year]:
    total_cases_by_month[year][month]  = 0
  if 'total_cases' in element:
    total_cases_by_month[year][month] = element['total_cases']

df_total_cases_by_month = pd.DataFrame.from_dict(total_cases_by_month)
df_total_cases_by_month
```

```python
df_total_cases_by_month.to_excel("USA_total_by_month.xlsx")
```

Zsumowane ilości lotów w 2019, 2020 i 2021 (brak 22 bo niepełny)

```python
amount_of_flights = {
    "amount_2019" : 0,
    "amount_2020" : 0,
    "amount_2021" : 0
}
for a in data_csv["2019"]:
  amount_of_flights['amount_2019'] += int(a.replace(',',''))
for a in data_csv["2020"]:
  amount_of_flights['amount_2020'] += int(a.replace(',',''))
for a in data_csv["2021"]:
  amount_of_flights['amount_2021'] += int(a.replace(',',''))  

print(amount_of_flights)
```

Zsumowane ilości lotów w poszczególnych miesiącach 2019,20,21,22

```python
from collections import OrderedDict

total_cases_by_month_year = {
  "2019" : {},
  "2020" : {},
  "2021" : {},
  "2022" : {},
}

for year in data_csv:
  if year is "Date":
     continue
  for i in range(len(data_csv[str(year)])):
    date = ""
    if "/" in data_csv["Date"][i]:
      date = data_csv["Date"][i].split("/")
    else:
      date = data_csv["Date"][i].split(".")
    date[0] = "{:02d}".format(int(date[0]))
    if date[0] not in total_cases_by_month_year["2019"]:
      total_cases_by_month_year["2019"][date[0]] = 0
    if date[0] not in total_cases_by_month_year["2020"]:
      total_cases_by_month_year["2020"][date[0]] = 0
    if date[0] not in total_cases_by_month_year["2021"]:
      total_cases_by_month_year["2021"][date[0]] = 0
    if date[0] not in total_cases_by_month_year["2022"]:
      total_cases_by_month_year["2022"][date[0]] = 0

    if data_csv["2022"][i]:
      try:
        total_cases_by_month_year["2022"][date[0]] += int(str(data_csv["2022"][i]).replace(",", ""))
      except:
        pass
    if data_csv["2021"][i]:
      total_cases_by_month_year["2021"][date[0]] += int(str(data_csv["2021"][i]).replace(",", ""))
    if data_csv["2020"][i]:
      total_cases_by_month_year["2020"][date[0]] += int(str(data_csv["2020"][i]).replace(",", ""))
    if data_csv["2019"][i]:
      total_cases_by_month_year["2019"][date[0]] += int(str(data_csv["2019"][i]).replace(",", ""))
    

#print(json.dumps(total_cases_by_month_year, indent=4, sort_keys=True))
total_cases_by_month_year = json.loads(json.dumps(total_cases_by_month_year, indent=4, sort_keys=True))
total_cases_by_month_year1 = pd.DataFrame.from_dict(total_cases_by_month_year)
total_cases_by_month_year1
```

**Cena w dolarach za galon paliwa lotniczego**

```python
jetfuel = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/jet fuel price.xlsx")
jetfuel_data = {}

for i in range(len(jetfuel["Month"])):
  #month = jetfuel["Month"][i].split()[0]
  #year = jetfuel["Month"][i].split()[1]
  date = jetfuel["Month"][i]
  price = jetfuel["Price"][i]
  change = jetfuel["Change"][i]
  if date not in jetfuel_data:
    jetfuel_data[date] = { "Price" : 0,
                          "Change": 0}

  jetfuel_data[date]["Price"] = price
  jetfuel_data[date]["Change"] = change

#print(json.dumps(jetfuel_data, indent=4))
jetfuel_pd = pd.DataFrame.from_dict(jetfuel_data)
jetfuel_pd
```
